```python
updatedData = data.drop(columns=["Date", "User"])
updatedData = updatedData.iloc[: , 1:]
updatedData
```

```python
def remove_links_mentions(tweet):
  link_re_pattern = "https?:\/\/t.co/[\w]+"
  mention_re_pattern = "@\w+"
  tweet = re.sub(link_re_pattern, "", tweet)
  tweet = re.sub(mention_re_pattern, "", tweet)
  return tweet.lower()
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
      u"\U0001F600-\U0001F64F" # emoticons
      u"\U0001F300-\U0001F5FF" # symbols & pictographs
      u"\U0001F680-\U0001F6FF" # transport & map symbols
      u"\U0001F1E0-\U0001F1FF" # flags (iOS)
      "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
tweets = updatedData['Tweets']=updatedData['Tweets'].apply(str)
tweets
```

```python
sentences = [word.lower() for word in tweets if re.match('^[a-zA-Z]+', word)]
sentences
```

```python
from nltk.tokenize import word_tokenize
# separating label(category) and tokenizing the tweet itself and based on that splitting data into training and testing
updatedTweets = [word_tokenize(remove_links_mentions(tweet)) for tweet in sentences]
```

```python
from nltk.corpus import stopwords
#Stop words removal
data = updatedTweets

tweetsList = []
for i in range (len(data)):
  data = updatedTweets[i]
  # data = [tweet for tweet in updatedTweets[0]] # a list comprehension
  data = [word for word in data if not word in set(stopwords.words('english'))]
  data = ' '.join(data)
  tweetsList.append(data)
  # print(data)

tweetsList[:1]
```

```python
from nltk.stem import PorterStemmer
#Stemming
stemmer=PorterStemmer()
for i in range(len(tweetsList)):
  words=word_tokenize(tweetsList[i])
  #List comprehension
  words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
  tweetsList[i]=' '.join(words)
print(tweetsList)
```

```python
from nltk.stem import WordNetLemmatizer
#Stemming with lemmatization to get proper meaning words after stemming
#Create obj of Lemmatizer
lemmmatizer=WordNetLemmatizer()

for i in range(len(tweetsList)):
    words=word_tokenize(tweetsList[i])
    #List comprehension
    words = [lemmmatizer.lemmatize(word.lower()) for word in words if word not in set(stopwords.words('english'))]
    tweetsList[i]=' '.join(words)
print(tweetsList)
```
